```python
#7 & 8 Traiter la colonne warranty afin d'uniformiser les valeurs (avoir uniquement des valeurs numÃ©riques uniformes)
#      Traiter la colonne warranty afin d'obtenir une colonne numÃ©rique avec 3 valeurs uniques

csv["warranty"] = csv["warranty"].str[0].astype(int)
csv
```

```python
#9 Traiter la colonne price afin d'avoir 3 catÃ©gories :  
#
#    - 0-300 
#
#    -301-500 
#
#    - 501- ++ 


csv.groupby(pd.cut(csv["price"], [0, 300,500,1000])).mean().loc[:,["capacity", "failure_rate", "margin","market_share", "attractiveness"]]
```

```python
#10 Transformer la colonne product_type en colonne numÃ©rique avec 3 catÃ©gories :  
#
#    1 
#    2 
#    3 

csv["product_type"] = csv["product_type"].apply(lambda x: 1 if x=="auto-portee" else 2 if x=="electrique" else 3)
```

```python
csv.groupby(["product_type"]).first()
```

```python
#11) Puis crÃ©er 3 colonnes (auto-portÃ©e, essence, Ã©lectrique) reprÃ©sentant une variable binaire avec 1 
#     si le produit est du type de la colonne ou 0 sinon. 

csv["auto-portee"]=0
csv["electrique"]=0
csv["essence"]=0

csv.loc[csv["product_type"]==1,"auto-portee"]=1
csv.loc[csv["product_type"]==2,"electrique"]=1
csv.loc[csv["product_type"]==3,"essence"]=1

csv.loc[:,['auto-portee' , 'electrique' ,'essence']]
```
